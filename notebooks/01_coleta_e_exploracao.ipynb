```python
# ============================================================
# Desertos Médicos SUS - Coleta e Exploração
# ============================================================

import sys
sys.path.append("..")

import pandas as pd
from src.config import (
    DATA_EXTERNAL, DATA_PROCESSED, OUTPUT_MAPAS,
    COMPETENCIA_DEFAULT, IBGE_SP_CAPITAL,
)
from src.data_loader import ElasticnesDownloader

print("✅ Ambiente configurado!")
print(f"📁 Dados externos: {DATA_EXTERNAL}")
print(f"📁 Processados: {DATA_PROCESSED}")
```

```python
# ============================================================
# Download de Dados do ElastiCNES
# ============================================================

downloader = ElasticnesDownloader(output_dir=DATA_EXTERNAL)

# Download de SP (ou use download_brasil() para todas as UFs)
df_profissionais = downloader.download_uf(
    uf="SP",
    competencia=COMPETENCIA_DEFAULT,
    force=False,  # True para forçar re-download
)

print(f" Total de registros: {len(df_profissionais):,}")
display(df_profissionais.head())
```

```python
# ============================================================
# Filtrar Apenas Médicos
# ============================================================

from src.analysis import filtrar_medicos, deduplicar_medicos_por_local

df_medicos = filtrar_medicos(df_profissionais)
print(f"👨‍⚕️ Total de vínculos de médicos: {len(df_medicos):,}")

df_medicos_unicos = deduplicar_medicos_por_local(df_medicos)
print(f"✅ Médicos únicos por estabelecimento: {len(df_medicos_unicos):,}")

# Salvar
df_medicos_unicos.to_parquet(DATA_PROCESSED / "medicos_sus_unicos.parquet", index=False)
```
